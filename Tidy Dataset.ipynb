```python
weather = pd.read_csv("data\\weather.csv")
```

```python
weather.head()
```

```python
weather_molten = pd.melt(weather,
                        id_vars=['id', 'year', 'month', 'element'],
                        var_name='day',
                        value_name='temperature')
weather_molten.head()
```

```python
 weather_tidy = weather_molten.pivot_table(
 index = ['id','year','month','day'],
 columns = 'element',
 values = 'temperature')
    
weather_tidy.head()
```

```python
weather_tidy.reset_index(inplace = True)

weather_tidy.head()
```

```python
weather_tidy['day'] = weather_tidy['day'].apply(lambda x: x[1:])

weather_tidy['date'] = weather_tidy['year'].astype(str) + '-' + weather_tidy['month'].astype(str) + '-' + weather_tidy['day'].astype(str)


weather_tidy.drop(['year','month','element','day'],axis =  1 ,inplace = True)
weather_tidy.head()
```
